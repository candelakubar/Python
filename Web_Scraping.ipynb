```python
pd.set_option('display.max_rows', None)


# Encontrar las filas que contienen la palabra 'Player'
filtro = df2014['Player'].str.contains('Player')

# Mostrar las filas que cumplen el filtro

# Eliminar las filas que cumplen el filtro
df2014 = df2014.drop(df2014[filtro].index)

# Mostrar el dataframe resultante
print(df2014)
#Reinicio indice
df2014 = df2014.reset_index(drop=True)
```

Acá lo que quiero ver es cuál de todos los equipos obtuvo mayor cantidad de puntos

```python

pts_tm=df2014[['PTS','Tm']].groupby(by = "Tm").sum().reset_index().sort_values('PTS',ascending=False)

pts_tm=pd.DataFrame(pts_tm)

#Elimino el Team 'TOT' que corresponde a el promedio de puntos hecho por los jugadores que jugaron en más de un equipo
pts_tm.drop([28], axis=0, inplace=True)

 
```

```python

f, axs = plt.subplots(figsize=(18, 4))

sns.set_theme(style="whitegrid")
ax = sns.barplot(x='Tm', y='PTS', data=pts_tm)
```

### PTS VS. MP

```python
sns.jointplot(x='PTS', y= 'MP', data=df2014,kind='reg'
             )
plt.show()

```

Queda confirmado que para encestar muchos puntos hay que jugar muchos minutos.


```python
sns.jointplot(x='PTS', y= '2P', data=df2014, kind='reg')
plt.show()
```

```python
sns.jointplot(x='PTS', y='2PA', data=df2014, kind='reg')
```

```python
sns.jointplot(x='PTS', y='FT', data=df2014, kind='reg')
```

Vemos que a menor cantidad de tiros libres, los puntos son menores, lo cual tiene sentido. 

## Podemos concluir con estos tres gráficos y analizando sus pendientes, que para obtener más puntos, es significativamente más importante tener más minutos de juego en la cancha.
